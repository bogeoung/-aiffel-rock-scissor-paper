```python
import os
import pandas as pd
import numpy as np
import warnings
from sklearn.ensemble import RandomForestRegressor
warnings.filterwarnings('ignore')
```

```python
os.listdir("./data")
```

```python
train_path = os.path.join("data", "train.csv")
test_path = os.path.join("data", "test.csv")
submission_path = os.path.join("data", "submission_sample.csv")
bts_path = os.path.join("data", "bus_bts.csv")
```

```python
train = pd.read_csv(train_path)
test = pd.read_csv(test_path)
submission = pd.read_csv(submission_path)
bts = pd.read_csv(bts_path)
```

```python
# train : 날짜, 출근시간의 버스 정류장별 승하차 인원, 퇴근시간 정류장별 승차인원
train.head()
```

```python
# test : 출근시간의 버스 정류장별 승하차 인원
test.head(10)
```

```python
submission.head()
```

```python
bts.head()
```

```python
set(train.columns) - set(test.columns)
```

```python
y_train = train["18~20_ride"]
```

```python
# train test 합치기
data = pd.concat([train.drop('18~20_ride', axis=1), test]).reset_index(drop = True)
```

```python
print(train.shape)
print(test.shape)
print(data.shape)
```

```python
data = data.drop(["station_name", "bus_route_id","id","date"], axis = 1)
data
```

```python
# null 값이 있는지 확인
data.isna().sum()
```

```python
for col in data.columns:
    print(f"column: {col}")
    print(f"The number of unique : {data[col].nunique()}")
    print()
```

```python
for col in data.columns:
    print(f"column: {col}")
    print(f"unique : {data[col].unique()}")
    print()
```

```python
data = pd.get_dummies(data, columns=["in_out"])
```

```python
data.shape
```

```python
data.head()
```

```python
len(train)
```

```python
X_train = data[:len(train)]
X_test = data[len(train):].reset_index(drop = True)
```

```python
params = {
    "n_estimators" : 300,
    "random_state" : 120
}
```

```python
rf = RandomForestRegressor(**params)
```

```python
y_train
```

```python
rf.fit(X_train, y_train)
```

```python
pred = rf.predict(X_test)
```

```python
pred = np.expm1(pred)
```

```python
submission['18~20_ride'] = pred
```

```python
submission.to_csv("./data/submission_baseline_rf.csv", index = False)
```
