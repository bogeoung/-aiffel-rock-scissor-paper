```python
import cv2
import numpy as np
import os
from glob import glob
from os.path import join
import tarfile
import urllib

from matplotlib import pyplot as plt
import tensorflow as tf
```

### 1. 사진 준비하기

```python
# 준비한 이미지 읽기
img_path = os.getenv('HOME')+'/aiffel/human_segmentation/images/my_image.jpeg'  # 본인이 선택한 이미지의 경로에 맞게 바꿔 주세요. 
img_orig = cv2.imread(img_path) 
print (img_orig.shape)
```

### 2. 세그멘테이션으로 사람 분리하기

```python
# DeepLab 모델로 DeepLabModel 클래스 만듬
class DeepLabModel(object):
    INPUT_TENSOR_NAME = 'ImageTensor:0'
    OUTPUT_TENSOR_NAME = 'SemanticPredictions:0'
    INPUT_SIZE = 513
    FROZEN_GRAPH_NAME = 'frozen_inference_graph'

    # __init__()에서 모델 구조를 직접 구현하는 대신, tar file에서 읽어들인 그래프구조 graph_def를 
    # tf.compat.v1.import_graph_def를 통해 불러들여 활용하게 됩니다. 
    def __init__(self, tarball_path):
        self.graph = tf.Graph()
        graph_def = None
        tar_file = tarfile.open(tarball_path)
        for tar_info in tar_file.getmembers():
            if self.FROZEN_GRAPH_NAME in os.path.basename(tar_info.name):
                file_handle = tar_file.extractfile(tar_info)
                graph_def = tf.compat.v1.GraphDef.FromString(file_handle.read())
                break
        tar_file.close()

        with self.graph.as_default():
    	    tf.compat.v1.import_graph_def(graph_def, name='')

        self.sess = tf.compat.v1.Session(graph=self.graph)

    # 이미지를 전처리하여 Tensorflow 입력으로 사용 가능한 shape의 Numpy Array로 변환합니다.
    def preprocess(self, img_orig):
        height, width = img_orig.shape[:2]
        resize_ratio = 1.0 * self.INPUT_SIZE / max(width, height)
        target_size = (int(resize_ratio * width), int(resize_ratio * height))
        resized_image = cv2.resize(img_orig, target_size)
        resized_rgb = cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)
        img_input = resized_rgb
        return img_input
        
    def run(self, image):
        img_input = self.preprocess(image)

        # Tensorflow V1에서는 model(input) 방식이 아니라 sess.run(feed_dict={input...}) 방식을 활용합니다.
        batch_seg_map = self.sess.run(
            self.OUTPUT_TENSOR_NAME,
            feed_dict={self.INPUT_TENSOR_NAME: [img_input]})

        seg_map = batch_seg_map[0]
        return cv2.cvtColor(img_input, cv2.COLOR_RGB2BGR), seg_map
```

```python
# 사전에 학습된 가중치를 불러옴
# define model and download & load pretrained weight
_DOWNLOAD_URL_PREFIX = 'http://download.tensorflow.org/models/'

model_dir = os.getenv('HOME')+'/aiffel/human_segmentation/models'
tf.io.gfile.makedirs(model_dir)

print ('temp directory:', model_dir)

download_path = os.path.join(model_dir, 'deeplab_model.tar.gz')
if not os.path.exists(download_path):
    urllib.request.urlretrieve(_DOWNLOAD_URL_PREFIX + 'deeplabv3_mnv2_pascal_train_aug_2018_01_29.tar.gz',
                   download_path)

MODEL = DeepLabModel(download_path)
print('model loaded successfully!')
```

```python
# 이미지를 네트워크에 입력
img_resized, seg_map = MODEL.run(img_orig)
print (img_orig.shape, img_resized.shape, seg_map.max())
```

```python
# 사람 영역만 검출
img_show = img_resized.copy()
seg_map = np.where(seg_map == 15, 15, 0) # 예측 중 사람만 추출
img_mask = seg_map * (255/seg_map.max()) # 255 normalization
img_mask = img_mask.astype(np.uint8)
color_mask = cv2.applyColorMap(img_mask, cv2.COLORMAP_JET)
img_show = cv2.addWeighted(img_show, 0.6, color_mask, 0.35, 0.0)

plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.show()
```

### 3. 세그멘테이션 결과를 원래 크기로 복원하기 

```python
img_mask_up = cv2.resize(img_mask, img_orig.shape[:2][::-1], interpolation=cv2.INTER_LINEAR)
_, img_mask_up = cv2.threshold(img_mask_up, 128, 255, cv2.THRESH_BINARY)

ax = plt.subplot(1,2,1)
plt.imshow(img_mask_up, cmap=plt.cm.binary_r)
ax.set_title('Original Size Mask')

ax = plt.subplot(1,2,2)
plt.imshow(img_mask, cmap=plt.cm.binary_r)
ax.set_title('DeepLab Model Mask')

plt.show()
```

### 4. 배경 흐리게 하기

```python
# 배경을 흐리게 하기 위해서 세그멘테이션 마스크를 이용해서 배경만 추출
img_mask_color = cv2.cvtColor(img_mask_up, cv2.COLOR_GRAY2BGR)
img_bg_mask = cv2.bitwise_not(img_mask_color)
img_bg = cv2.bitwise_and(img_orig, img_bg_mask)
plt.imshow(img_bg)
plt.show()
```

```python
img_bg_blur = cv2.blur(img_bg, (13,13))
plt.imshow(cv2.cvtColor(img_bg_blur, cv2.COLOR_BGR2RGB))
plt.show()
```

### 5. 흐린 배경과 원본 영상 합성

```python
img_concat = np.where(img_mask_color==255, img_orig, img_bg_blur)
plt.imshow(cv2.cvtColor(img_concat, cv2.COLOR_BGR2RGB))
plt.show()
```

### 6. 함수화 시키기

```python
def print_shape(img_path):
    img_orig = cv2.imread(img_path)
    #print(img_orig.shape)
    return img_orig

def get_weight(img_path,segmap,imgbackground):
    # 사전에 학습된 가중치를 불러옴
    # define model and download & load pretrained weight
    _DOWNLOAD_URL_PREFIX = 'http://download.tensorflow.org/models/'

    model_dir = os.getenv('HOME') + '/aiffel/human_segmentation/models'
    tf.io.gfile.makedirs(model_dir)

    #print('temp directory:', model_dir)

    download_path = os.path.join(model_dir, 'deeplab_model.tar.gz')
    if not os.path.exists(download_path):
        urllib.request.urlretrieve(_DOWNLOAD_URL_PREFIX + 'deeplabv3_mnv2_pascal_train_aug_2018_01_29.tar.gz',
                                   download_path)

    MODEL = DeepLabModel(download_path)
    #print('model loaded successfully!')

    img_orig = print_shape(img_path)
    # 이미지를 네트워크에 입력
    img_resized, seg_map = MODEL.run(img_orig)
    #print(img_orig.shape, img_resized.shape, seg_map.max())

    img_back_orig = print_shape(imgbackground)
    img_background_resized, seg_map2 = MODEL.run(img_back_orig)
    #print(img_back_orig.shape, img_resized.shape, seg_map2.max())

    # 사람 영역만 검출
    #img_show = img_resized.copy()
    seg_map = np.where(seg_map == segmap, segmap, 0)  # 예측 중 사람만 추출
    img_mask = seg_map * (255 / seg_map.max())  # 255 normalization
    img_mask = img_mask.astype(np.uint8)
    color_mask = cv2.applyColorMap(img_mask, cv2.COLORMAP_JET)
    #img_show = cv2.addWeighted(img_show, 0.6, color_mask, 0.35, 0.0)
    #plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
    #plt.show()

    #img_show2 = img_back_orig.copy()
    seg_map2 = np.where(seg_map2 == segmap, segmap, 0)
    img_background = seg_map2 * (255 / seg_map2.max())
    img_background = img_background.astype(np.uint8)
    #img_show2 = cv2.addWeighted(img_show2, 0.6, color_mask, 0.35, 0.0)
    #plt.imshow(cv2.cvtColor(img_show2, cv2.COLOR_BGR2RGB))
    #plt.show()


    # 세그멘테이션 결과를 원래 크기로 복원
    img_mask_up = cv2.resize(img_mask, img_orig.shape[:2][::-1], interpolation=cv2.INTER_LINEAR)
    _, img_mask_up = cv2.threshold(img_mask_up, 128, 255, cv2.THRESH_BINARY)

    img_background_up = cv2.resize(img_background, img_back_orig.shape[:2][::-1], interpolation=cv2.INTER_LINEAR)
    _, img_background_up = cv2.threshold(img_background_up, 128, 255, cv2.THRESH_BINARY)


    #ax = plt.subplot(1, 2, 1)
    #plt.imshow(img_background_up, cmap=plt.cm.binary_r)
    #ax.set_title('Original Size Mask')

    #ax = plt.subplot(1, 2, 2)
    #plt.imshow(img_background, cmap=plt.cm.binary_r)
    #ax.set_title('DeepLab Model Mask')
    #plt.show()


    # 배경을 흐리게 하기 위해서 세그멘테이션 마스크를 이용해서 배경만 추출
    img_mask_color = cv2.cvtColor(img_mask_up, cv2.COLOR_GRAY2BGR)
    img_bg_mask = cv2.bitwise_not(img_mask_color)
    img_bg = cv2.bitwise_and(img_back_orig, img_bg_mask)
    img_bg_blur = cv2.blur(img_bg, (13, 13))
    #plt.imshow(img_bg_blur)
    #plt.show()

    # 흐린 배경과 원본 영상 합성
    img_concat = np.where(img_mask_color == 255, img_orig, img_bg_blur)
    plt.imshow(cv2.cvtColor(img_concat, cv2.COLOR_BGR2RGB))
    plt.show()
```

### 7. 인물모드 사진에서의 문제점과 개선할 수 있는 솔루션은?

```python
from IPython.display import Image

img_path = os.getenv('HOME')+'/aiffel/human_segmentation/images/problem.jpg'
Image(img_path)
```

인물모드 사진에서의 문제점은 다양하지만 위 사진에 대해서는 다음 2개의 문제점이 발견되었다.

1. 여러 사람이 있을 시 모든 사람 추출
 <br/> 한혜진만 블러 처리가 되질 않길 원했지만, 사진에 있는 모든 사람(심지어 왼쪽 팔 사이어 얼굴만 있는 사람까지도)이 블러처리가 되지 않는다. <br/> 
2. 부자연스러운 경계점
 <br/> 한혜진의 오른쪽 팔을 보면 팔과 몸통 사이의 배경 부분에 대해서 경계점이 생성되지 않았다.


위와 같은 문제점은 어떻게 해결할까?<br/>
참고1 : https://www.jeremyjordan.me/semantic-segmentation/
<br/>
참고2 : https://blog.lunit.io/2018/07/02/deeplab-v3-encoder-decoder-with-atrous-separable-convolution-for-semantic-image-segmentation/

현재 우리가 사용한 방법은 물리적 의미 단위로 인식하는 Semantic Segmentation이다.<br/>
픽셀의 카테고리만을 분류하고, 각각의 물체가 무엇인지 구분하지 않는다. <br/>
즉 같은 범주의 클래스가 두 개 있는 경우, 이를 별도의 개체로 구분하지 않아,
사람이 누구인지 관계없이 같은 라벨로 표현이 된다는 의미이다.<br/>

그렇기 때문에 위 사진에서 한혜진과 행인은 다른 사람이지만 `같은 사람 클래스`로써 구분이 된다.<br/>
그 결과 DeepML의 세그멘테이션을 수행하면, 같은 15(사람 라벨)의 값을 갖게 된다.<br/>
마스크를 만들고 시각화 할 때 DeepML의 라벨을 기준으로 분류를 하기 때문에, 사람이기만 하면 블러의 대상에서 제외가 된다. <br/>

이를 해결하기 위해서는 사람 개개인 별로 다른 라벨을 가지게 하는 `instance segmentation`을 사용해야만 할 것이다.<br/>
instance segmentation을 사용하면 `한혜진과 행인은 각각 다른 클래스`로써 구분이 될 것이다.<br/>
segmentation을 수행하고 나서도 서로 다른 라벨을 가지고 있을 것이기 때문에, 라벨을 기준으로 마스크를 만들면 한혜진만 블러처리가 되지 않도록 할 수 있을 것이다.
